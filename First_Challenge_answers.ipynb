```python
#q17 : write a function whihc will take multiple list as a input and give me concatnation of all the element as an output
def list_concat(my_list):
    concated_list=[]
    for i in my_list:
        if type(i)== int or type(i)== str or type(i)== float:
            concated_list.append(i)
        if type(i)== list or type(i)== tuple:
            for j in i:
                concated_list.append(j)
    return concated_list
```

```python
list_concat([1,2,34,4.0,5,[4,4,4,322,"g"],2344,43,4,24,2,(7,4,3,2)])
```
